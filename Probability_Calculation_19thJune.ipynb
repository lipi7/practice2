```python
import numpy as np
import random



#Function for rolling the dice
def roll_the_dice(n_simulations = 100000):
    
    count = 0
    #score =0
    for i in range(n_simulations):
            
            die1 = random.randint(1,6)
            die2 = random.randint(1,6)
            
            score = die1 + die2
            
            if((score%2 ==0)or (score > 7)):
                count+=1
    return count/n_simulations

string = 'The probability of rolling an even or reater than 7 is: '
print(string,np.round(roll_the_dice()*100,2),'%')
```

```python
# 12 transistors type  A=12 and type B=18

import random
trials = 10000
totalA = 12
totalB = 18

successes = 0

choicelist = list("A" * totalA + "B" * totalB)

print('Choicelist: '+choicelist)
def transistor():
    return random.choice(choicelist)
for i in range(trials):
    try1 = transistor()
    try2 = transistor()
    if try1 == "A" and try2 == "B":
        successes +=1
print(float(successes/trials))
```

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
plt.ylabel('Probablity of cars passing')
plt.xlabel('Number of cars')
plt.title('Probability Distribution Curve')
arr =[]
rv = poisson(25)
for num in range(0,40):
    arr.append(rv.pmf(num))
    
prob = rv.pmf(28)
print(prob)
plt.grid(True)
plt.plot(arr,linewidth=2.0)
plt.plot([28],[prob],marker='o',markersize=6,color='red')
plt.show()
```
